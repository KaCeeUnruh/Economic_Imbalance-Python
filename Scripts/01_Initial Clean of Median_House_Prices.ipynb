## Transpose Data

```python
df_transposed =df_house3.melt(id_vars=['State'], var_name='Year', value_name='House Cost')
```

```python
df_transposed
```

```python
# Replace missing values with the mean from nonmissing values within the same year and state
def replace_zero_with_mean(group):
    mean_val = group[group != 0].mean()
    group[group == 0] = mean_val
    return group

df_transposed['House Cost'] = df_transposed.groupby(['Year', 'State'])['House Cost'].transform(replace_zero_with_mean)
```

```python
df_transposed
```

```python
# Merge States within each year
df_final = df_transposed.groupby(['Year', 'State'], as_index=False)['House Cost'].mean()
df_final
```

```python
# Address the missing Value in SD 2008: Take value from SD 2009 and put it in place to fina closest estimation
sd_2009_row = df_final[(df_final['Year'] == '2009') & (df_final['State'] == 'SD')]

if not sd_2009_row.empty:
    value_sd_2009 = sd_2009_row['House Cost'].iloc[0]
    df_final.loc[(df_final['State'] == 'SD') & (df_final['Year'] == '2008'), 'House Cost'] = value_sd_2009

df_final.head(40)
```

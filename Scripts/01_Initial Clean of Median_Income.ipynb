## Transpose Data

```python
# Transposed Data to have 3 columns: State, Year, Income
df_transposed =df_clean_2.melt(id_vars=['State'], var_name='Year', value_name='Income')
```

```python
df_transposed
```

```python
# Make a new order for the COlumns
new_order = ['Year', 'State', 'Income']
```

```python
# Reorder the columns
df_transposed = df_transposed[new_order]
df_transposed
```

```python
df_transposed = df_transposed.sort_values(by=['Year', 'State'])
df_transposed
```

```python
df_transposed = df_transposed.reset_index(drop=True)
df_transposed
```
